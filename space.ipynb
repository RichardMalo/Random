```python
import pygame
import random

# Constants for the game
WIDTH, HEIGHT = 800, 600
SHIP_WIDTH, SHIP_HEIGHT = 50, 50
ITEM_WIDTH, ITEM_HEIGHT = 30, 30
FPS = 60

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))

# Load images
ship_img = pygame.image.load('ship.png')
item_img = pygame.image.load('item.png')

# Game loop
running = True
score = 0
ship_pos = [WIDTH // 2, HEIGHT // 2]
item_pos = [random.randrange(WIDTH), random.randrange(HEIGHT)]
while running:
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Ship movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        ship_pos[0] -= 5
    if keys[pygame.K_RIGHT]:
        ship_pos[0] += 5
    if keys[pygame.K_UP]:
        ship_pos[1] -= 5
    if keys[pygame.K_DOWN]:
        ship_pos[1] += 5

    # Collision detection
    ship_rect = pygame.Rect(ship_pos[0], ship_pos[1], SHIP_WIDTH, SHIP_HEIGHT)
    item_rect = pygame.Rect(item_pos[0], item_pos[1], ITEM_WIDTH, ITEM_HEIGHT)
    if ship_rect.colliderect(item_rect):
        score += 1
        item_pos = [random.randrange(WIDTH), random.randrange(HEIGHT)]

    # Drawing
    screen.fill((0, 0, 0))
    screen.blit(ship_img, ship_pos)
    screen.blit(item_img, item_pos)
    pygame.display.flip()

    # Cap the framerate
    pygame.time.Clock().tick(FPS)

pygame.quit()
```
